```python
#Imports
import pandas as pd
import missingno as msno
```

```python
#Read in csv and create dataframe
NAm_metadata = pd.read_csv('data/NAm_metadata.csv')
#Display 5 rows for reference
NAm_metadata.head(5)
```

```python
#Summarize dataframe 
NAm_metadata.info()
```

```python
#Print number of "NaN" values by column
missing_values = NAm_metadata.isnull().sum()
print(missing_values)
```

```python
#Create Matrix showing location of "NaN" values
msno.matrix(NAm_metadata, labels='True')
```

```python
#Print number of "?" values by column
question_marks = (NAm_metadata == '?').sum()

print(question_marks)
```

```python
#Display bar chart of number of "NaN" per column
msno.bar(NAm_metadata, sort="ascending")
```

```python
#Replace "?" with "NaN"
NAm_no_qm = NAm_metadata.replace('?', pd.NA)
```

```python
#Display matrix of "NaN" values, including values that were previously "?"
msno.matrix(NAm_no_qm, labels='True')
```
