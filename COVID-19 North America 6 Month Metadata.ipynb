```python
import pandas as pd
```

```python
#Read in .tsv file, specifying tab delimiter
cov_na_6m = pd.read_csv("nextstrain_ncov_open_north-america_6m_metadata.tsv", sep='\t')
print(cov_na_6m.info())
```
